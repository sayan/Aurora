Trapping Rain Water
You are given an array non-negative integers height which represent an elevation map. Each value height[i] represents the height of a bar, which has a width of 1.

Return the maximum area of water that can be trapped between the bars.



Input: height = [0,2,0,3,1,0,1,3,2,1]

Output: 9
Constraints:

1 <= height.length <= 1000
0 <= height[i] <= 1000

```python
def trap(height: List[int]) -> int:
    len_heights = len(height)
    l_index = 0
    possible_water_trapped = []
    while l_index < len_heights - 1: ## inequality will ensure if heights are of len 5 it 
        right_index = l_index + 1
        this_bucket_total = 0
        this_bucket_land_mass = 0
        ## we will search for the next partition. 
        while (right_index < len_heights) and (height[l_index] > height[right_index]):
            print(f"l index {l_index}, right index {right_index}")
            this_bucket_land_mass += height[right_index]
            right_index += 1
            if right_index == len_heights:
                print(f"between {l_index} and {right_index} we cant store water")
                l_index = right_index -1
        total_water_area = (right_index - l_index -1) * height[l_index]
        print(f"Between {l_index} and {right_index} total water area is {total_water_area}")
        print(f"Between {l_index} and {right_index} total land area is {this_bucket_land_mass}")
        total_actual_water_area = 0 if total_water_area <= 0 else total_water_area - this_bucket_land_mass
        possible_water_trapped.append(total_actual_water_area)
        l_index = right_index
    return possible_water_trapped


    
```

```python
from typing import List
def at_required_positon(l:List[int]) -> List[int]:
    even_pos = 0
    odd_pos = 1

    while even_pos < len(l):
        if l[even_pos] % 2 == 1:
            print(f"Swapping {even_pos} with next odd pos {odd_pos}")
            l[even_pos], l [odd_pos] = l[odd_pos], l[even_pos]
            odd_pos = odd_pos +2
        else:
            even_pos = even_pos + 2
    return(l)

```
